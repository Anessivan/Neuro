```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t=list()
x=list()
x_=list()
y=list()
y_=list()

with open("Ensemble.txt") as f:
    for line in f:
        _t, _x, _x_, _y, _y_ = map(float, line.split())
        t.append(_t)
        x.append(_x)
        x_.append(_x_)
        y.append(_y)
        y_.append(_y_)
```

```python
plt.plot(t, x)
```

```python
plt.plot(t, x_)
```

```python
plt.plot(t, y)
```

```python
plt.plot(t, y_)
```
